```python
header_2 = exempledf.iloc[tabs_index[1]]
sec_part.columns = header_2
sec_part = sec_part.iloc[:,:sec_part.columns.get_loc("Domicile")]
```

```python
sec_part = sec_part.drop(columns=COLLS_NO_CAT)
```

```python
map_ = dict()
for col in sec_part.columns:
    map_[col] = dict()
    for line in sec_part[col]:
        if type(line) not in [int,float,np.nan]:
            splt = line.split('.') if "." in line[:round(len(line)/2)] else line.split(' ')
            index = splt[0]
            map_[col][index] = line.replace(index,'').replace('.','').strip()
```

```python
map_['Dem.2'] = map_['Dem.3'] = map_['Dem.1']
```
